```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-8,8,0.1)
w1 = 0.5
w2 = 1.0
w3 = 2.5

l1 = 'Weight w = 0.5'
l2 = 'Weight w = 1.0'
l3 = 'Weight w = 2.0'

for w, l in [(w1,l1),(w2,l2),(w3,l3)]:
    f = 1/(1+np.exp(-x*w))
    plt.plot(x,f,label=l)
plt.legend(loc=4)
plt.grid()
plt.show()
```

```python
b1 = -3
b2 = 0
b3 = 2

l1 = 'Delay b = -3'
l2 = 'Delay b = 0'
l3 = 'Delay b = 2'

for b, l in [(b1,l1),(b2,l2),(b3,l3)]:
    f = (1/(1+np.exp(-x+b*1)))
    plt.plot(x,f,label=l)

plt.legend(loc=4)
plt.grid()
plt.show()
```
